In this exercise, we will use the lemmatization process to produce the proper form of a given word.

1. Open a Jupyter colab or Notebook.

2. Insert a new cell and add the following code to import the necessary libraries:

```python
from nltk import download
download('wordnet')
download('omw-1.4')
from nltk.stem.wordnet import WordNetLemmatizer
```

3. Create an object of the **WordNetLemmatizer** class.

```python
lemmatizer = WordNetLemmatizer()
```

4. Bring the word to its proper form by using the **lemmatize()** method of the **WordNetLemmatizer** class.

```python
def get_lemma(word):
  return lemmatizer.lemmatize(word)

get_lemma('products')
```

5. Similarly, use the input as production now:

```python
get_lemma('production')
```

6. Similarly, use the input as coming now:

```python
get_lemma('coming')
```
