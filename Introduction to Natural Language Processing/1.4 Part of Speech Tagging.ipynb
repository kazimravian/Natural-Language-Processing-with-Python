<ul>
<font color = 'brown green'>

<li>
In NLP, the term PoS refers to parts of speech.
</li><br>

<li>
PoS tagging refers to the process of tagging words within sentences with their respective PoS.
</li><br>

<li>
We extract the PoS of tokens constituting a sentence so that we can filter out the PoS that are of interest and analyze them.
</li><br>

<li>
For example, if we look at the sentence, "The sky is blue," we get
four tokens, namely "The," "sky," "is," and "blue", with the help of tokenization.
</li><br>

<li>
Now, using a PoS tagger, we tag the PoS for each word/token. This will look as follows:<br>
[('The', 'DT'), ('sky', 'NN'), ('is', 'VBZ'), ('blue', 'JJ')]
</li><br>

<li>
DT = Determiner <br>
NN = Noun, common, singular or mass<br>
VBZ = Verb, present tense, third-person singular<br>
JJ = Adjective<br>
</li><br>

</ul>

In this exercise, we will find out the PoS for each word in the sentence, **I am
reading NLP Fundamentals**. We first make use of tokenization in order to get
the tokens. Later, we will use the **pos_tag()** method, which will help us find the PoS for each word/token. Follow these steps to implement this exercise:

1. Open a Jupyter Notebook.

2. Insert a new cell and add the following code to import the necessary libraries:

```python
from nltk import word_tokenize, pos_tag
```

3. To find the tokens in the sentence, we make use of the **word_tokenize()**
method. Insert a new cell and add the following code to implement this:

```python
def get_tokens(sentence):
  words = word_tokenize(sentence)
  return words
```

4. Print the tokens with the help of the **print()** function. To implement this, add a new cell and write the following code:

```python
words = get_tokens("I am reading NLP Fundamentals")
words
```

5. We shall now use the **pos_tag()** method. Insert a new cell and add the
following code:

```python
def get_pos(words):
  return pos_tag(words)

get_pos(words)
```
