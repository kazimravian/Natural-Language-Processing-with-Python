In this exercise, we will normalize some given text. Basically, we will be trying to replace select words with new words, using the **replace()** function, and finally produce the normalized text. <br>
**replace()** is a built-in Python function that works on strings and takes two arguments. It will return a copy of a string in which the occurrence of the first argument will be replaced by the second argument.

1. Open a Jupyter or colab Notebook.

2. Insert a new cell and add the following code to assign a string to the **sentence** variable:

```python
sentence = "I visited the US from the UK on 22-10-18"
```

3. We want to replace **"US"** with **"United States"**, **"UK"** with **"United
Kingdom"**, and **"18"** with **"2018"**. To do so, use the **replace()** function and store the updated output in the **"normalized_sentence"** variable.

```python
def normalize(text):
  return text.replace("US", "United States").replace("UK", "United Kingdom").replace("-18", "-2018")
```

4. To check whether the text has been normalized, insert a new cell and add the
following code to print it:

```python
normalized_sentence = normalize(sentence)
print(normalized_sentence)
```

5. Add the following code:

```python
normalized_text = normalize('US and UK are two superpowers')
print(normalized_text)
```
