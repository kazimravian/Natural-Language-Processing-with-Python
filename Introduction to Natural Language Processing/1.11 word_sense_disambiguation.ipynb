<li>
A word's meaning depends on its association with other words in a sentence.
</li><br>

<li>
This means two or more words with the same spelling may have different meanings
in different contexts. This often leads to ambiguity.
</li><br>

<li>
Word sense disambiguation is the process of mapping a word to the sense that it should carry.
</li><br>

<li>
We need to disambiguate words based on the sense they carry so that they can be treated as different entities when being analyzed.
</li><br>

<li>
The following figure displays a perfect example of how ambiguity is caused due to the usage of the same word in different sentences:
</li><br>

<h2>
<b>

<ul>
<font color = 'brown green'>

<li>
One of the algorithms to solve word sense disambiguation is the Lesk algorithm.
</li><br>

<li>
It has a huge corpus in the background (generally WordNet is used) that contains
definitions of all the possible synonyms of all the possible words in a language.
</li><br>

<li>
Then it takes a word and the context as input and finds a match between the context and all the definitions of the word.
</li><br>

<li>
The meaning with the highest number of matches with the context of the word will be returned.
</li><br>

</ul>

In this exercise, we will find the sense of the word **"bank"** in two different sentences.

1. Open a Jupyter or colab Notebook.

2. Insert a new cell and add the following code to import the necessary libraries:

```python
import nltk
from nltk.wsd import lesk
from nltk import word_tokenize
nltk.download('omw-1.4')
nltk.download('wordnet')
nltk.download(['punkt'])
```

3. Declare two variables, **sentence1** and **sentence2**, and assign them with
appropriate strings.

```python
sentence1 = "Keep your savings in the bank"
sentence2 = "It's so risky to drive over the banks of the road"
```

4. To find the sense of the word **"bank"** in the preceding two sentences, use the **Lesk algorithm** provided by the **nltk.wsd** library.

```python
def get_synset(sentence, word):
  return lesk(word_tokenize(sentence), word)
```

```python
get_synset(sentence1, 'bank')
```

Here, **savings_bank.n.02** refers to a container for keeping money safely at
home.

```python
get_synset(sentence2, 'bank')
```

Here, **bank.v.07** refers to a slope in the turn of a road.
