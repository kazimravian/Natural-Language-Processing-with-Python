In this exercise, we will pass a few words through the stemming process so that they get converted into their base forms.

1. Open a Jupyter or colab Notebook.

2. Insert a new cell and add the following code to import the necessary libraries:

```python
from nltk import stem
```

3. Now pass the following words as parameters to the **stem()** method.

```python
def get_stems(word, stemmer):
  return stemmer.stem(word)

porterStem = stem.PorterStemmer()
```

4. When the input is **"production"**, the following output is generated:

```python
get_stems("production", porterStem)
```

5. Similarly, the following code would be used for the input **"coming"**. We get the following output:

```python
get_stems("coming", porterStem)
```

6. Similarly, the following code would be used for the input **"firing"**. When the input is **"firing"**, the following output is generated:

```python
get_stems("firing", porterStem)
```

7. The following code would be used for the input **"battling"**. If we give the input "battling", the following output is generated:

```python
get_stems("battling", porterStem)
```

8. The following code will also generate the same output as above, for the
input **"battling"**. The output will be as follows:

```python
stemmer = stem.SnowballStemmer("english")

get_stems("battling", stemmer)
```
