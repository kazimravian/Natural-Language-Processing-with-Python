The Bag of Words (BoW) model is one of the most popular methods for extracting
features from raw texts.
</li><br>

<li>
In this technique, we convert each sentence into a vector. The length of this vector is equal to the number of unique words in all the documents. This is done in two steps:

<ul>
<font color = 'white' size = "5">
<br>
<li>
The vocabulary or dictionary of all the words is generated.
</li><br>

<li>
The document is represented in terms of the presence or absence of all words.
</li>
</font>
</ul>
</li><br>

<li>
A vocabulary or dictionary is created from all the unique possible words available in the corpus (all documents) and every single word is assigned a unique index number.
</li><br>

<li>
In the second step, every document is represented by a list whose length is equal to the number of words in the vocabulary.
</li><br>

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

3. Use the **CountVectorizer** class to create the BoW model. Add the
following code to do this:

```python
def vectorize_text(corpus):

  bag_of_words_model = CountVectorizer()

  dense_vec_matrix = bag_of_words_model.fit_transform(corpus).todense()

  bag_of_words_df = pd.DataFrame(dense_vec_matrix)

  bag_of_words_df.columns = sorted(bag_of_words_model.vocabulary_)

  return bag_of_words_df
```

```python
corpus = ['Data Science is an overlap between Arts and Science',
          'Generally, Arts graduates are right-brained and '
          'Science graduates are left-brained',
          'Excelling in both Arts and Science at a time '
          'becomes difficult',
          'Natural Language Processing is a part of Data Science'
          ]
```

```python
df = vectorize_text(corpus)
df
```

4. Create a BoW model for the 10 most frequent terms. Add the following code to
implement this:

```python
def bow_top_n(corpus, n):

  bag_of_words_model_small = CountVectorizer(max_features=n)

  bag_of_word_df_small = pd.DataFrame(bag_of_words_model_small.fit_transform(corpus).todense())

  bag_of_word_df_small.columns = sorted(bag_of_words_model_small.vocabulary_)

  return bag_of_word_df_small
```

```python
df_2 = bow_top_n(corpus, 10)
df_2.head()
```
