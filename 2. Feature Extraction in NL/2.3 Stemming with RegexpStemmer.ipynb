<h1><b><font color = 'brown'>
Stemming
</font></b></h1>

<h2>
<b>

<ul>
<font color = 'brown green'>

<li>
In many languages, the base forms of words change when they're used in sentences.
</li><br>

<li>
For example, the word "produce" can be written as "production" or "produced" or
even "producing," depending on the context.
</li><br>

<li>
The process of converting a word back into its base form is known as stemming.
</li><br>

<li>
It is essential to do this, because without it, algorithms would treat two or more different forms of the same word as different entities, despite them having the same semantic meaning.
</li><br>

<li>
So, the words "producing" and "produced" would be treated as different entities, which can lead to erroneous inferences.
</li><br>

<li>
In Python, RegexpStemmer and PorterStemmer are the most widely used stemmers. Let's explore them one at a time.
</li><br>

</font>
</ul>
</b>
</h2>

<h1><b><font color = 'brown'>
RegexpStemmer
</font></b></h1>

<h2>
<b>

<ul>
<font color = 'brown green'>

<li>
RegexpStemmer uses regular expressions to check whether morphological or
structural prefixes or suffixes are present.
</li><br>

<li>
For instance, in many cases, verbs in the present continuous tense (the present tense form ending with "ing") can be restored to their base form simply by removing "ing" from the end; for example, "playing" becomes "play".
</li><br>

</font>
</ul>
</b>
</h2>

<h1><b><font color = 'brown'>
Exercise 1: Converting Words in the Present Continuous Tense into Base
Words with RegexpStemmer
</font></b></h1>

In this exercise, we will use **RegexpStemmer** on text to convert words into their basic form by removing some generic suffixes such as "ing" and "ed".
To use **nltk's regex_stemmer**, we have to create an object of **RegexpStemmer** by passing the regex of the suffix or prefix and an integer, **min**, which indicates the minimum length of the stemmed string.

1. Open a Jupyter or Colab Notebook.

2. Insert a new cell and import **RegexpStemmer**:

```python
from nltk.stem import RegexpStemmer
```

3. Use **regex_stemmer** to stem each word of the **sentence** variable.

```python
def get_stems(text):
  """
  Creating an object of RegexpStemmer, any string ending
  with the given regex 'ing$' will be removed.
  """
  regex_stemmer = RegexpStemmer('ing$', min = 4)

  """
  The below code line will convert every word into its
  stem using regex stemmer and then join them with space.
  """
  return ' '.join([regex_stemmer.stem(wd) for wd in text.split()])

sentence = "I love playing football"

get_stems(sentence)  
```

<h1><b><font color = 'brown'>
The Porter Stemmer
</font></b></h1>

<h2>
<b>

<ul>
<font color = 'brown green'>

<li>
The Porter stemmer is the most common stemmer for dealing with English words.
</li><br>

<li>
It removes various morphological and inflectional endings (such as suffixes, prefixes, and the plural "s") from English words.
</li><br>

<li>
In doing so, it helps us extract the base form of a word from its variations.
</li><br>

</font>
</ul>
</b>
</h2>

<h1><b><font color = 'brown'>
Exercise 2: Using the Porter Stemmer
</font></b></h1>

In this exercise, we will apply the Porter stemmer to some text. Follow these steps to complete this exercise:

1. Open a Jupyter or Colab Notebook.

2. Import **nltk** and any related packages and declare a **sentence** variable.

```python
from nltk.stem.porter import *

sentence = "Before eating, it would be nice to sanitize your hands with a sanitizer"
```

3. Now, we'll make use of the Porter stemmer to stem each word of the **sentence** variables:

```python
def get_stems(text):
  ps_stemmer = PorterStemmer()

  return ' '.join([ps_stemmer.stem(wd) for wd in text.split()])

get_stems(sentence)
```
