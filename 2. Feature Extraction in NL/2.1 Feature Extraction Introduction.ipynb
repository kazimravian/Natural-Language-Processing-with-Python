<h1><b><font color = 'brown'>
Categorizing Data Based on Content
</font></b></h1>

<h2>
<b>

<ul>
<font color = 'brown green'>

<li>
Data can be divided into four categories based on content, as shown in the
following diagram:
</li><br>

</ul>
</b>
</h2>

image.png

<h2>
<b>

<ul>
<font color = 'brown green'>

<li>
Let's look at each category here:

<ul>
<font color = 'brown'>
<li>
<font color = 'brown green'>Text data:</font><br>
This refers to text corpora consisting of written sentences. This type of data can only be read. An example would be the text corpus of a book.
</li>
<li>
<font color = 'brown green'>Image data:</font><br>
This refers to pictures that are used to communicate messages.
This type of data can only be seen.
</li>
<li>
<font color = 'brown green'>Audio data:</font><br>
This refers to voice recordings, music, and so on. This type of data can only be heard.
</li>
<li>
<font color = 'brown green'>Video data:</font><br>
A continuous series of images coupled with audio forms a video.
This type of data can be seen as well as heard.
</li>
</font>
</ul>

</li><br>

</ul>
