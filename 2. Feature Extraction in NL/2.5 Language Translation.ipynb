Exercise 01: Language Translation
</font></b></h1>

In this exercise, we will make use of the **TextBlob** library to translate a sentence from Spanish into English.

1. Open a Jupyter or Colab Notebook.

2. Import **TextBlob**, as follows:

```python
from textblob import TextBlob
```

3. Make use of the **translate()** function of **TextBlob** to translate the input text from Spanish to English.

```python
def translate(text, from_1, to_1):
  en_blob = TextBlob(text)
  return en_blob.translate(from_1, to_1)

# traslate spanish into english
translate(text = 'muy bien', from_1 = 'es', to_1 = 'en')
```

<h1><b><font color = 'brown'>
Exercise 02: Removing Stop Words from Text
</font></b></h1>

In this exercise, we will remove the stop words from a given text.

1. Open a Jupyter Notebook.

2. Import **nltk** and declare a **sentence** variable with the text in question:

```python
import nltk
nltk.download('punkt')
from nltk import word_tokenize
```

```python
sentence = "She sells seashells on the seashore"
sentence
```

3. Define a **remove_stop_words** method and remove the custom list of stop
words from the **sentence** by using the following lines of code:

```python
def remove_stop_words(text, stop_word_list):
  return ' '.join([word for word in word_tokenize(text) if word.lower() not in stop_word_list])

custom_stop_word_list = ['she', 'on', 'the', 'am', 'is', 'not']
remove_stop_words(sentence, custom_stop_word_list)
```
