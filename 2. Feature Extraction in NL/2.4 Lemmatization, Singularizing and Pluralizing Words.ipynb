<h1><b><font color = 'brown'>
Exercise 01: Performing Lemmatization
</font></b></h1>

In this exercise, we will perform lemmatization on some text. Follow these steps to complete this exercise:

1. Open a Jupyter or Colab Notebook.

2. Import **nltk** and its related packages, and then declare a **sentence** variable.

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
nltk.download('omw-1.4')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
sentence = "The products produced by the process today are far better than what it produces generally."
sentence
```

3. To lemmatize the tokens, we extracted from the sentence:

```python
lemmatizer = WordNetLemmatizer()

def get_lemmas(text):
  
  return ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(text)])

get_lemmas(sentence)
```

<h1><b><font color = 'brown'>
Exercise 02: Singularizing and Pluralizing Words 
</font></b></h1>

In this exercise, we will make use of the **textblob** library to singularize and pluralize words in the given text.

1. Open a Jupyter or Colab Notebook.

2. Import **TextBlob** and declare a **sentence** variable.

```python
from textblob import TextBlob

sentence = TextBlob('She sells seashells on the seashore')
print(sentence)
print(sentence.words)
```

3. To singularize the third word in the sentence, type the following code:

```python
def singularize(word):
  return word.singularize()

singularize(sentence.words[2]) 
```

4. To pluralize the fifth word in the given sentence, type the following code:

```python
def pluralize(word):
  return word.pluralize()

pluralize(sentence.words[5])
```
